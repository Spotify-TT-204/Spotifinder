```python
import pandas as pd
import numpy as np
```

```python
tracks_df = pd.read_csv('./tracks.csv.zip')
```

```python
display(tracks_df.shape)
tracks_df.release_date = tracks_df.release_date.astype('datetime64')
display(tracks_df.head())
display(tracks_df.release_date.describe())
mod_tracks_df = tracks_df[tracks_df.release_date >= '1984']
display(mod_tracks_df.shape)
display(mod_tracks_df.head())
display(mod_tracks_df.release_date.describe())
```

```python
display(mod_tracks_df.isna().sum())

# 3 rows with null values for the 'name' column - filter them out
mod_tracks_df = mod_tracks_df[mod_tracks_df.name.isna() == False]

display(mod_tracks_df.isna().sum())
```

```python
match_artist = lambda x: [x in artists for artists in mod_tracks_df.artists]
match_song = lambda x: [len(set(x.lower().split()).intersection(set(sng.lower().split()))) >= len(x.split())  for sng in mod_tracks_df.name]

display(mod_tracks_df[match_artist('Bruno Mars')])
display(mod_tracks_df[match_song("Don't worry be happy")])
```

```python
spotify_df = pd.read_csv('./spotify_songs.csv.zip')
display(spotify_df.shape)
spotify_df.head()
```

```python
spotify_en_df = spotify_df[spotify_df.language == 'en']
display(spotify_en_df.shape)
spotify_en_df.head()
```

```python
display(spotify_en_df.isna().sum())
```

```python
spotify_en_df.track_artist.describe()
```

```python
spotify_en_df.track_album_release_date.sort_values()
```

```python
match_artist = lambda x: [x in artists for artists in spotify_en_df.track_artist]
match_song = lambda x: [len(set(x.lower().split()).intersection(set(sng.lower().split()))) >= len(x.split())  for sng in spotify_en_df.track_name]
match_lyric = lambda x: [len(set(x.lower().split()).intersection(set(sng.lower().split()))) >= len(x.split())  for sng in spotify_en_df.lyrics]
```

```python
display(spotify_en_df[match_artist('McFerrin')])
display(spotify_en_df[match_song("Don't worry be happy")])
display(spotify_en_df[match_lyric("fast car")])
```
